```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_fiber_optic(angle_of_incidence_degrees, core_radius, length):
    angle_rad = np.radians(angle_of_incidence_degrees)

    x_positions = [0]
    y_positions = [0]

    current_x = 0
    current_y = 0
    direction_y = 1

    x_step = core_radius / np.tan(angle_rad)

    while current_x < length:
        current_x += x_step
        current_y += (core_radius * direction_y)

        if current_x > length:
            break

        x_positions.append(current_x)
        y_positions.append(current_y)

        direction_y *= -1

    return x_positions, y_positions

# --- Візуалізація ---
cable_length = 50
cable_radius = 2
# Градус нахилу повинен бути менше 90 градусів.
angle = 50

x, y = simulate_fiber_optic(angle, cable_radius, cable_length)

plt.figure(figsize=(10, 3))
plt.plot(x, y, color='red', label='Промінь світла')


plt.axhline(cable_radius, color='blue', linestyle='--', label='Стінки кабелю')
plt.axhline(0, color='gray', linestyle=':', label='Center Axis')
plt.axhline(-cable_radius, color='blue', linestyle='--')

plt.xlim(0, cable_length)
plt.title("Симуляція променів у оптоволоконному кабелі")
plt.legend(loc='upper right')
plt.show()
```
